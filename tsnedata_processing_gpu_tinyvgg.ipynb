```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Hyper-parameters 

class_labels = 8



conv_row,conv_col = 10,29
out_channel = 1
# 16x128

# 
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.conv1 = nn.Conv2d(1, 1, (3, 3), (1, 2))
        self.conv2 = nn.Conv2d(1, 1, (3, 3), (1, 2))
        self.conv3 = nn.Conv2d(1, 1, (3, 3), (1, 1))
        
        self.fc1 = nn.Linear(conv_row * conv_col * out_channel, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 8)
        
        self.dropout = nn.Dropout(0.5)  # Dropout with a probability of 0.5

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        
        x = x.view(-1, conv_col * conv_row * out_channel)           
        
        x = F.relu(self.fc1(x))
        x = self.dropout(x)  # Applying dropout regularization
        x = F.relu(self.fc2(x))
        x = self.dropout(x)  # Applying dropout regularization
        x = self.fc3(x)
        x = F.softmax(x, dim=1)
        return x

# test_samplesx8
def testAccuracy(model,test_loader):  
    with torch.no_grad():
        n_correct = 0
        n_samples = 0
        class_labels =8
        n_class_correct = [0 for i in range(class_labels)]
        n_class_samples = [0 for i in range(class_labels)]
        classes = [i for i in range(class_labels)]
    
        for images, labels in test_loader:

            #images = images.FloatTensor()
            
            images = images.unsqueeze(1) 
            #print(images.shape)
            images = images.to(device).float()
            labels = labels.to(device)
            outputs = model(images)
            # flattened_output = outputs.reshape(outputs.shape[0], -1)
            # print(flattened_output.shape)




            #print(np.array(outputs).shape)
            # max returns (value ,index)
            _, predicted = torch.max(outputs, 1)
            #print(np.array(predicted).shape)
            #print(labels.size(0))
            n_samples += labels.size(0)
            n_correct += (predicted == labels).sum().item()

            for i in range(len(labels)):
                label = labels[i]
                pred = predicted[i]
                if (label == pred):
                    n_class_correct[label] += 1
                n_class_samples[label] += 1
        
        
      

        total_acc = 100.0 * n_correct / n_samples
        print(f'Accuracy of the network: {total_acc} %')


        # for i in range(class_labels):
        #     acc = 100.0 * n_class_correct[i] / n_class_samples[i]
        #     print(f'Accuracy of {classes[i]}: {acc} %')
        
        return total_acc

    
def modelTrainer(train_loader,test_loader):
    
    num_epochs = 10
    
    learning_rate = 0.001

    class_labels = 8

   

    # get some random training images
    dataiter = iter(train_loader)
    images, labels = next(dataiter)

    model = ConvNet().to(device)
    print(model)

    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)


    bestaccuracy = 0
    bestepoch = 0

    accList = []
    lossList = []

    n_total_steps = len(train_loader)
    for epoch in range(num_epochs):
        for images, labels in train_loader:
            

            images = images.unsqueeze(1) 
            #print(images.shape)
            #images = images.reshape(64,1,256,63)
            images = images.to(device).float()
            labels = labels.to(device)
            
            

            # Forward pass
            outputs = model(images)
            

            
            # print(outputs)
            # print(labels)
            
            loss = criterion(outputs, labels)

            # Backward and optimize
            # optimizer.zero_grad()
            # loss.backward()
            # optimizer.step()
            
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
        
       
        accuracy = testAccuracy(model,test_loader)

        print (f'Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{n_total_steps}], Loss: {loss.item():.4f},Accuracy: {accuracy} ')
        
        accList.append(accuracy)
        lossList.append(loss.item())

        if(bestaccuracy < accuracy):
            bestaccuracy = accuracy
            bestepoch = epoch
            PATH = './cnn.pth'
            torch.save(model, PATH)


    print('Finished Training')
  

    print(bestaccuracy)
    print(bestepoch)
    
    return accList,lossList



print("7")
print("train lablight and test lablight")
accList_7_lab,lossList_7_lab = modelTrainer(train_loader,test_loader)
```
